```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define the total units and days in the year
total_units = 2500
days_in_year = 365

# Calculate the mean (lambda) for the Poisson distribution
mean_lambda = total_units / days_in_year

# Generate a Poisson distribution
data_poisson = poisson.rvs(mu=mean_lambda, size=days_in_year)

# Plot the distribution
plt.hist(data_poisson, bins=30, density=True, alpha=0.6, color='g')

# Plot the PMF
xmin, xmax = plt.xlim()
x = np.arange(xmin, xmax)
plt.plot(x, poisson.pmf(x, mean_lambda), 'bo', ms=8, label='poisson pmf')
plt.show()

# Print the mean of the distribution
print("Mean of the distribution: ", np.mean(data_poisson))
```
